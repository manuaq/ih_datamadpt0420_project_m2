```python
diamonds_df['cut'].unique()
```

```python
diamond_cut_ordered_list = ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair']
print(len(diamond_cut_ordered_list))
print(diamonds_df['cut'].unique())
print(len(diamonds_df['cut'].unique()))
print('diamond_cut_ordered_list', diamond_cut_ordered_list)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x=diamonds_df['cut'], data=diamonds_df, order=diamond_cut_ordered_list, palette="Blues")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='center', fontweight='light',fontsize='medium')
plt.xlabel("From left: Ideal (Best Cut) to Fair Cut", fontweight='bold', size=13)
plt.ylabel("Diamond Count", fontweight='bold', size=13)
plt.title("Available Diamonds by Cut", size=16, fontweight='bold')

plt.show()
```

```python
diamonds_df['color'].unique()
```

```python
diamond_color_ordered_list = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
print(len(diamond_color_ordered_list))
print(diamonds_df['color'].unique())
print(len(diamonds_df['color'].unique()))
print('diamond_color_ordered_list', diamond_color_ordered_list)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x=diamonds_df['color'], data=diamonds_df, order=diamond_color_ordered_list, palette="Blues")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='center', fontweight='light',fontsize='medium')
plt.xlabel("From left: D (Best Colorless) to J (Near Colorless)", fontweight='bold', size=13)
plt.ylabel("Diamond Count", fontweight='bold', size=13)
plt.title("Available Diamonds by Color", size=16, fontweight='bold')

plt.show()
```

```python
sns.catplot(x="color", y="price", kind="boxen", data=diamonds_df.sort_values("color"))
plt.gcf().set_size_inches(12, 5)
plt.xlabel("Color Grades: Best from left -> D (Colorless) to J (Near Colorless)", fontweight='bold', size=13)
plt.ylabel("Diamond Prices", fontweight='bold', size=13)
plt.title("Diamond Prices by Color", size=16, fontweight='bold')
plt.show();
```

```python
sns.boxplot(x="color", y="price", data=diamonds_df.sort_values("color"))
plt.gcf().set_size_inches(12, 5)
plt.xlabel("Color Grades: Best from left -> D (Colorless) to J (Near Colorless)", fontweight='bold', size=13)
plt.ylabel("Diamond Prices", fontweight='bold', size=13)
plt.title("Diamond Prices by Color", size=16, fontweight='bold')
plt.show();
```

```python
plt.figure(figsize=[12,12])

# First subplot showing the diamond carat weight distribution
plt.subplot(221)
plt.hist(diamonds_df['carat'],bins=20,color='b')
plt.xlabel('Carat Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Carat Weight')

# Second subplot showing the diamond depth distribution
plt.subplot(222)
plt.hist(diamonds_df['depth'],bins=20,color='r')
plt.xlabel('Diamond Depth (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Depth')

# Third subplot showing the diamond price distribution
plt.subplot(223)
plt.hist(diamonds_df['price'],bins=10,color='g')
plt.xlabel('Price in USD')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Price')

plt.show()
```

```python
plt.figure(figsize=(12, 12))
correlation = diamonds_df.select_dtypes(include = ['float64', 'int64']).iloc[:, 1:].corr()
sns.heatmap(correlation, vmax=1, annot=True,square=True)
plt.show()
```

```python
diamonds_df['cut_num'] = diamonds_df['cut'].apply(lambda x: 1 if x=='Fair' else(2 if x=='Good' 
                                           else(3 if x=='Very Good' 
                                           else(4 if x=='Premium' else 5))))

diamonds_df['color_num'] = diamonds_df['color'].apply(lambda x: 1 if x=='J' else(2 if x=='I'
                                          else(3 if x=='H'
                                          else(4 if x=='G'
                                          else(5 if x=='F'
                                          else(6 if x=='E' else 7))))))

diamonds_df['clarity_num'] = diamonds_df['clarity'].apply(lambda x: 1 if x=='I1' else(2 if x=='SI2'
                                          else(3 if x=='SI1'
                                          else(4 if x=='VS2'
                                          else(5 if x=='VS1'
                                          else(6 if x=='WS2'
                                          else 7 if x=='WS1' else 8))))))

diamonds_df
```
